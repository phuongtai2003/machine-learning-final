```python
#521H0480 - Nguyen Phuong Tai
#521H0288 - Tran Hoang Phuc
import pandas as pd

# Loading the dataset
file_path = './titanic.csv'
titanic_data = pd.read_csv(file_path)

# Previewing dataset
titanic_data.head()

# Features Listing.
# PassengerId: An identification of passengers.
# Survived: Survival result 0 for no, 1 for yes.
# Pclass: The class of passenger including 1,2,3.
# Name: Passenger name.
# Sex: Passenger gender.
# Age: Passenger age.
# SibSp: Number of siblings or spouses aboard.
# Parch: Number of parents or children aboard.
# Ticket: Ticket number.
# Fare: The fare paid for the ticket.
# Cabin: Cabin number.
# Embarked: Port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton).
```

```python
# Statiscal and categorical summary
statistical_summary = titanic_data.describe()

categorical_summary = titanic_data.describe(include=['O'])

# Getting the summary data
statistical_summary, categorical_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Checking to find missing data
missing_data = titanic_data.isnull().sum().sort_values(ascending=False)

# Plot the missing data on a graph
plt.figure(figsize=(10,6))
missing_data.plot(kind='bar')
plt.title('Missing Data in the Dataset')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.show()

# Visualizing the distribution of numerical features
plt.figure(figsize=(15, 10))

# Age Distribution
plt.subplot(2, 3, 1)
sns.histplot(titanic_data['Age'].dropna(), kde=True)
plt.title('Age Distribution')

# Fare Distribution
plt.subplot(2, 3, 2)
sns.histplot(titanic_data['Fare'], kde=True)
plt.title('Fare Distribution')

# PClass Distribution
plt.subplot(2, 3, 3)
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Passenger Class Distribution')

# Sex Distribution
plt.subplot(2, 3, 4)
sns.countplot(x='Sex', data=titanic_data)
plt.title('Gender Distribution')

# Siblings/sprouses Distribution
plt.subplot(2, 3, 5)
sns.countplot(x='SibSp', data=titanic_data)
plt.title('Siblings/Spouses Aboard')

# Parents/Childs Distribution
plt.subplot(2, 3, 6)
sns.countplot(x='Parch', data=titanic_data)
plt.title('Parents/Children Aboard')

plt.tight_layout()
plt.show()

# Visualizing the relationship between features and survival
plt.figure(figsize=(20, 6))

# Survival Rate by Gender
plt.subplot(1, 5, 1)
sns.barplot(x='Sex', y='Survived', data=titanic_data)
plt.title('Survival Rate by Gender')

# Survival Rate by Passenger Class
plt.subplot(1, 5, 2)
sns.barplot(x='Pclass', y='Survived', data=titanic_data)
plt.title('Survival Rate by Passenger Class')

# Survival Rate by Embarkation Port
plt.subplot(1, 5, 3)
sns.barplot(x='Embarked', y='Survived', data=titanic_data)
plt.title('Survival Rate by Embarkation Port')

# Survival Rate by Age
plt.subplot(1, 5, 4)
sns.histplot(titanic_data, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title('Survival Rate by Age')

# Survival Rate by Fare
plt.subplot(1, 5, 5)
sns.histplot(titanic_data, x='Fare', hue='Survived', multiple='stack', bins=30)
plt.title('Survival Rate by Fare')

plt.tight_layout()
plt.show()
```

```python
# Handle missing data
# Convert cabin to cabin_indicator. If the data point has cabin then cabin_indicator = 1 else equal to 0
titanic_data['Cabin_Indicator'] = titanic_data['Cabin'].apply(lambda x: 0 if pd.isna(x) else 1)

# Fill missing age value, the age value take the median value group with passenger class and gender
median_ages = titanic_data.groupby(['Pclass', 'Sex'])['Age'].median()
titanic_data['Age'] = titanic_data.apply(lambda row: median_ages[row['Pclass'], row['Sex']] if pd.isna(row['Age']) else row['Age'], axis=1)

# Fill the embarked data with the most presented data
most_common_embarked = titanic_data['Embarked'].mode()[0]
titanic_data['Embarked'].fillna(most_common_embarked, inplace=True)

# Rechecking the data set
titanic_data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Dropping non relevant columns like passengerId, name, ticket and cabin
titanic_prepared = titanic_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

# Convert categorical data to numerical data, example: male, female -> 0, 1 
label_encoder = LabelEncoder()
titanic_prepared['Sex'] = label_encoder.fit_transform(titanic_prepared['Sex'])
titanic_prepared['Embarked'] = label_encoder.fit_transform(titanic_prepared['Embarked'])

# Splitting features and target
X = titanic_prepared.drop('Survived', axis=1)
y = titanic_prepared['Survived']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Checking the prepared data
X_train.head(), y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler

# Use basic machine learning models to predict
logistic_regression_model = LogisticRegression(max_iter=1000)
decision_tree_model = DecisionTreeClassifier(random_state=42)
random_forest_model = RandomForestClassifier(random_state=42)
gradient_boosting_model = GradientBoostingClassifier(random_state=42)
knn_model = KNeighborsClassifier()
naive_bayes_model = GaussianNB()

# Stores models and names
models = {
    "Logistic Regression": logistic_regression_model,
    "Decision Tree": decision_tree_model,
    "Random Forest": random_forest_model,
    "Gradient Boosting": gradient_boosting_model,
    "KNN Model": knn_model,
    "Naive Bayes": naive_bayes_model
}

# Train and evaluate model
def train_and_evaluate(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    confusion = confusion_matrix(y_test, y_pred)
    return accuracy, report, confusion

# Train and evaluate each model
results = {}
for model_name, model in models.items():
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    if(model_name == 'KNN Model' or model_name == 'Logistic Regression'):
        # Uses scaled data for KNN and Logistic Regression
        accuracy, report, confusion = train_and_evaluate(model, X_train_scaled, y_train, X_test_scaled, y_test)
    else:
        accuracy, report, confusion = train_and_evaluate(model, X_train, y_train, X_test, y_test)
    results[model_name] = {"Accuracy": accuracy, "Classification Report": report, "Confusion Matrix": confusion}
```

```python
# Display the model with the corresponding accuracy and confusion matrix data
simplified_results = {
    "Model": [],
    "Accuracy": [],
    "True Negative (TN)": [],
    "False Positive (FP)": [],
    "False Negative (FN)": [],
    "True Positive (TP)": []
}

for model_name, model_result in results.items():
    simplified_results["Model"].append(model_name)
    simplified_results["Accuracy"].append(model_result["Accuracy"])
    tn, fp, fn, tp = model_result["Confusion Matrix"].ravel()
    simplified_results["True Negative (TN)"].append(tn)
    simplified_results["False Positive (FP)"].append(fp)
    simplified_results["False Negative (FN)"].append(fn)
    simplified_results["True Positive (TP)"].append(tp)

# Converting to DataFrame for better readability
simplified_results_df = pd.DataFrame(simplified_results)
simplified_results_df.set_index("Model", inplace=True)
simplified_results_df
```

```python
# Printing classification report of each model
for model_name, model_result in results.items():
    print(model_name)
    print(model_result["Classification Report"])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU
from tensorflow.keras.optimizers import Adam

# Feed Forward Neural Network
ffnn_model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compiling Feed Forward Neural Network model
ffnn_model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Train the ffnn model
ffnn_history = ffnn_model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=0)

# Evaluating the FFNN model
ffnn_loss, ffnn_accuracy = ffnn_model.evaluate(X_test_scaled, y_test, verbose=0)

print("Feed Forward Neural Network Model Accuracy:",ffnn_accuracy, ", Feed Forward Neural Network Model Loss: ", ffnn_loss)

# Building the Recurrent Neural Network (RNN) model with LSTM
rnn_model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])),
    LSTM(25),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Reshaping input data for LSTM layer
X_train_scaled_reshaped = X_train_scaled.reshape((X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_scaled_reshaped = X_test_scaled.reshape((X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# Compiling the RNN model
rnn_model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Training the RNN model
rnn_history = rnn_model.fit(X_train_scaled_reshaped, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=0)

# Evaluating the RNN model
rnn_loss, rnn_accuracy = rnn_model.evaluate(X_test_scaled_reshaped, y_test, verbose=0)

print("Recurrent Neural Network Accuracy:", rnn_accuracy, ", Recurrent Neural Network Loss:", rnn_loss)
```

```python
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Adding dropout and regularization
ffnn_model = Sequential([
    Dense(64, activation='relu', kernel_regularizer='l2', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.5),
    Dense(32, activation='relu', kernel_regularizer='l2'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10)

# Compiling the FFNN model
ffnn_model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Training with early stopping
ffnn_history = ffnn_model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping], verbose=0)

ffnn_loss, ffnn_accuracy = ffnn_model.evaluate(X_test_scaled, y_test, verbose=0)

print("Feed Forward Neural Network Model Accuracy:",ffnn_accuracy, ", Feed Forward Neural Network Model Loss: ", ffnn_loss)


rnn_model = Sequential([
    LSTM(50, return_sequences=True, dropout=0.2, recurrent_dropout=0.2, input_shape=(1, X_train_scaled.shape[1])),
    LSTM(25, dropout=0.2, recurrent_dropout=0.2),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10)

# Compiling the RNN model
rnn_model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Training with early stopping
rnn_history = rnn_model.fit(X_train_scaled_reshaped, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping], verbose=0)

# Evaluating the RNN model
rnn_loss, rnn_accuracy = rnn_model.evaluate(X_test_scaled_reshaped, y_test, verbose=0)
print("Recurrent Neural Network Accuracy:", rnn_accuracy, ", Recurrent Neural Network Loss:", rnn_loss)
```

```python
from sklearn.model_selection import GridSearchCV

# Parameters grid for Gradient Boosting
param_grid_gb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 0.9, 1.0]
}

# Initialize GridSearchCV
grid_search_gb = GridSearchCV(estimator=GradientBoostingClassifier(random_state=42), 
                              param_grid=param_grid_gb, 
                              cv=5, 
                              n_jobs=-1, 
                              verbose=1, 
                              scoring='accuracy')

# Fit GridSearchCV
grid_search_gb.fit(X_train, y_train)
print("Best parameters for Gradient Boosting:", grid_search_gb.best_params_)
print("Best score for Gradient Boosting:", grid_search_gb.best_score_)

best_gb_model = grid_search_gb.best_estimator_
y_pred_gb = best_gb_model.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
print("Test set accuracy for Gradient Boosting:", accuracy_gb)
```

```python
# Parameters grid for Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV
grid_search_rf = GridSearchCV(estimator=RandomForestClassifier(random_state=42), 
                              param_grid=param_grid_rf, 
                              cv=5, 
                              n_jobs=-1, 
                              verbose=1, 
                              scoring='accuracy')

# Fit GridSearchCV
grid_search_rf.fit(X_train, y_train)

# Best parameters and best score
print("Best parameters for Random Forest:", grid_search_rf.best_params_)
print("Best score for Random Forest:", grid_search_rf.best_score_)

# Evaluate on the test set
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Test set accuracy for Random Forest:", accuracy_rf)
```
